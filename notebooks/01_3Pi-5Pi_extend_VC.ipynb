## Path
**VC = valid construction (GOS path)**

This notebook extends the valid-construction path in the triplet-phase-lock framework.


## Triplet

$$\Pi^{(1)} = (3\Pi, 4\Pi, 5\Pi)$$

- **3Pi** → bilateral extension
- **4Pi** → constraint transition (C)
- **5Pi** → domain-consistent ratio (D)


## ABCD overlay (valid construction)

- **A = anchor**
- **B = bilateral**
- **C = constraint**
- **D = domain consistency**

For this VC notebook, the emphasis shifts to **C** and **D**:

- constraint: $\frac{25 - 1}{25}$
- domain consistency: $\frac{24}{25} = 0.96$


## Core examples

Constrained ratio:

$$\frac{25 - 1}{25} = \frac{24}{25}$$

Decimal consistency:

$$\frac{24}{25} = 0.96$$

These are grouped as a progression from bilateral structure into constrained, domain-consistent ratio.


## Contrast (IA)

- $0 \mapsto 0.96$
- $2 = 0.96$


```python
# Minimal VC examples for Triplet 1
ratio = (25 - 1) / 25
decimal = 24 / 25

print('Constrained ratio:', ratio)
print('Decimal form:', decimal)
print('Domain-consistent equality:', ratio == decimal)
```

## Interpretation

This notebook studies how valid construction extends by introducing:

1. a **constraint** on a defined set
2. a **ratio** produced by constrained removal
3. **domain consistency** between fractional and decimal form

In later VC notebooks, this path extends to larger stable systems.
