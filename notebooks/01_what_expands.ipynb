# 01 — what expands?

**triplet-phase-lock**

Pi stage: what expands?

```python
import numpy as np
import matplotlib.pyplot as plt

from src.expand import sequence_n, build_triplets_from_values

plt.rcParams['figure.figsize'] = (8, 4)
plt.rcParams['axes.grid'] = True
```

## Setup

We begin with the base sequence

\[
N_k = 24k - 25
\]

and triplets

\[
T_k = (N_k, N_{k+1}, N_{k+2})
\]

```python
num_steps = 40
k = np.arange(1, num_steps + 3)

values = sequence_n(k)
triplets = build_triplets_from_values(values)

print('First 10 sequence values:')
print(values[:10].astype(int))

print('\nFirst 5 triplets:')
print(triplets[:5].astype(int))

print('\nTriplet array shape:')
print(triplets.shape)
```

```python
plt.figure()
plt.plot(k, values, marker='o')
plt.xlabel('k')
plt.ylabel('N_k')
plt.title(r'Raw sequence: $N_k = 24k - 25$')
plt.tight_layout()
plt.show()
```

```python
triplet_index = np.arange(1, len(triplets) + 1)

plt.figure()
plt.plot(triplet_index, triplets[:, 0], marker='o', label=r'$N_k$')
plt.plot(triplet_index, triplets[:, 1], marker='o', label=r'$N_{k+1}$')
plt.plot(triplet_index, triplets[:, 2], marker='o', label=r'$N_{k+2}$')
plt.xlabel('triplet index')
plt.ylabel('value')
plt.title('Triplet coordinates across k')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
gaps = np.diff(values)

print('Unique local gaps:', np.unique(gaps))

plt.figure()
plt.plot(np.arange(1, len(gaps) + 1), gaps, marker='o')
plt.xlabel('k')
plt.ylabel(r'$N_{k+1} - N_k$')
plt.title('Local expansion gaps')
plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.plot(triplets[:, 0], triplets[:, 1], marker='o')
plt.xlabel(r'$N_k$')
plt.ylabel(r'$N_{k+1}$')
plt.title('2D projection of triplet candidates')
plt.tight_layout()
plt.show()
```
