# 01 — what expands?

**triplet-phase-lock**

This notebook starts the repo by defining a minimal expandable sequence, building triplet states, and visualizing the raw structure before any extension or resistance stage.

Guiding question:

- **Pi (expand): what expands?**


## Setup

We begin with a simple arithmetic seed:

\\[\nN_k = 24k - 25\n\\]\n
and package local structure into triplets:

\\[\nT_k = (N_k, N_{k+1}, N_{k+2})\n\\]\n
This notebook does **not** apply the 45° cosine gate yet. It only generates and visualizes candidate structure.


```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['figure.figsize'] = (8, 4)
plt.rcParams['axes.grid'] = True
```

## Pi — expand

Define the base sequence and a helper that packages neighboring values into triplets.


```python
def sequence_n(k):
    """
    Base sequence:
        N_k = 24k - 25
    Accepts a scalar or NumPy array.
    """
    k = np.asarray(k, dtype=float)
    return 24.0 * k - 25.0


def build_triplets(num_steps):
    """
    Build triplets:
        T_k = (N_k, N_{k+1}, N_{k+2})
    for k = 1, ..., num_steps.
    """
    ks = np.arange(1, num_steps + 3, dtype=float)
    vals = sequence_n(ks)
    return np.stack([vals[:-2], vals[1:-1], vals[2:]], axis=1)
```

## Generate candidate states


```python
num_steps = 40

k = np.arange(1, num_steps + 1)
n_vals = sequence_n(k)
triplets = build_triplets(num_steps)

print('First 10 sequence values:')
print(n_vals[:10].astype(int))

print('\nFirst 5 triplets:')
print(triplets[:5].astype(int))

print('\nTriplet array shape:')
print(triplets.shape)
```

## Plot 1 — raw sequence

This is the first expansion view: the arithmetic candidate sequence itself.


```python
plt.figure()
plt.plot(k, n_vals, marker='o')
plt.xlabel('k')
plt.ylabel('N_k')
plt.title(r'Raw sequence: $N_k = 24k - 25$')
plt.tight_layout()
plt.show()
```

## Plot 2 — triplet coordinates

Each row is now a local triplet state:

\\[\nT_k = (N_k, N_{k+1}, N_{k+2})\n\\]\n
Plot the three coordinates against the triplet index.


```python
triplet_index = np.arange(1, len(triplets) + 1)

plt.figure()
plt.plot(triplet_index, triplets[:, 0], marker='o', label=r'$N_k$')
plt.plot(triplet_index, triplets[:, 1], marker='o', label=r'$N_{k+1}$')
plt.plot(triplet_index, triplets[:, 2], marker='o', label=r'$N_{k+2}$')
plt.xlabel('triplet index')
plt.ylabel('value')
plt.title('Triplet coordinates across k')
plt.legend()
plt.tight_layout()
plt.show()
```

## Plot 3 — local expansion gaps

Even before any resistance stage, we can inspect how the sequence expands locally.

For this seed, consecutive gaps should be constant.


```python
gaps = np.diff(n_vals)

print('Unique local gaps:', np.unique(gaps))

plt.figure()
plt.plot(np.arange(1, len(gaps) + 1), gaps, marker='o')
plt.xlabel('k')
plt.ylabel(r'$N_{k+1} - N_k$')
plt.title('Local expansion gaps')
plt.tight_layout()
plt.show()
```

## Plot 4 — 2D triplet projection

A simple projection of the triplet state space using the first two coordinates.

This is a raw geometric view before introducing cosine alignment.


```python
plt.figure()
plt.plot(triplets[:, 0], triplets[:, 1], marker='o')
plt.xlabel(r'$N_k$')
plt.ylabel(r'$N_{k+1}$')
plt.title('2D projection of triplet candidates')
plt.tight_layout()
plt.show()
```
