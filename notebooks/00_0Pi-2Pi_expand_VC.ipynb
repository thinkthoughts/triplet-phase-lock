## Path
**VC = valid construction**

This notebook introduces the first valid-construction path in the triplet-phase-lock framework.


## Triplet

$$\Pi^{(0)} = (0\Pi, 1\Pi, 2\Pi)$$

- **0Pi** → domain / initial tokens
- **1Pi** → anchor
- **2Pi** → bilateral construction


## ABCD overlay

- **A = anchor**
- **B = bilateral**
- **C = constraint**
- **D = domain consistency**

For this first VC notebook, the emphasis is on **A** and **B**:

- anchor: $(x_0, t_0)$
- bilateral: $2 = 1 + 1$


## Core examples

Anchor:

$$(x_0, t_0)$$

Bilateral identity:

$$2 = 1 + 1$$

These are valid because they preserve domain and structure.


```python
# Minimal VC examples for Triplet 0
anchor = ('x0', 't0')
lhs = 2
rhs = 1 + 1

print('Anchor:', anchor)
print('Bilateral check:', lhs, '=', rhs)
print('Valid construction:', lhs == rhs)
```

## Interpretation

This notebook studies how a construction becomes valid by introducing:

1. an **anchor** $(x_0, t_0)$
2. a **bilateral** identity-preserving relation

In later VC notebooks, this path can be extended with:

- **C = constraint**
- **D = domain consistency**
