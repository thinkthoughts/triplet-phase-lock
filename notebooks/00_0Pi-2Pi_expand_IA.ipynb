## Path
**IA = invalid assignment (ZOS path)**

This notebook introduces the first invalid-assignment path in the triplet-phase-lock framework.


## Triplet

$$\Pi^{(0)} = (0\Pi, 1\Pi, 2\Pi)$$

- **0Pi** → unconstrained / undefined token space (Z0)
- **1Pi** → collapse transition (Z1)
- **2Pi** → assignment transition (Z2)


## ZABCD overlay (invalid states)

- **Z0 = unconstrained**
- **Z1 = collapse**
- **Z2 = assignment**
- **Z3 = denial**
- **Z4 = mixed inconsistency**

For this first IA notebook, the emphasis is on **Z0**, **Z1**, and **Z2**:

- unconstrained: $t$
- collapse: $1 - 1 = 0$
- assignment: $0 \mapsto 0.96$


## Core examples

Unconstrained token:

$$t$$

Collapse:

$$1 - 1 = 0$$

Assignment:

$$0 \mapsto 0.96$$

These are grouped as a progression from undefined state to unconstrained reassignment.

```python
# Minimal IA examples for Triplet 0
token = 't'
collapsed = 1 - 1
assigned = 0.96

print('Unconstrained token:', token)
print('Collapse:', '1 - 1 =', collapsed)
print('Assignment:', f'{collapsed} -> {assigned}')
```
